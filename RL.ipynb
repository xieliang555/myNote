# Basic Concepts
## Reward
- A reward $R_t$ is a scalar feedback signal  
- Indicates how well agent is doing at time step $t$  
- The agent's job is to maximise cumulative reward

## Decision Making
- Goal: select actions $A_t$ to maximise total future reward  
- Actions may have long term consequences  
- Reward may be delayed  
- It may be batter to sacrifice immediate reward to gain more long term reward  

## Observastion and History
- Recievie observation $O_t$ from the enviroment  
- History $H_t$ is the sequence of observations ,actons, and rewards $$H_t= O_1,R_1,A_1,...A_t-1,O_t,R_t$$

## History and State
- State $S_t$ is the information used to determine what happened next.   
- It's the information used by reinforcement learning algorithms. It can be any function of history: $$S_t=f(H_t)$$
- A state $S_t$ is Markov if and only if : $$P[S_t+1|S_t]=P[S_t+1|S_1,...S_t]$$
Once the state is known, the history may be thron away  

- Fully observability : $$S_t=O_t$$
- Partially observability : $$S_t=H_t$$

# Major Components of an RL Agent

## Policy
- agent's behaviour function
- it's a map from state to action
- deterministic policy: $a=\pi(s)$

## Value function
- value function is a prediction of future reward
- used to evalute the goodness/badness of states
- and therefore to select between actions e.g.
$$v_\pi(s)=\mathbb{E}_\pi[R_{t+1}+\gamma R_{t+2}+\gamma ^2 R_{t+3}+...|S_t=s]$$

## Model 
- a model predicts what the enviroment will do
- $\mathcal{P}$ predicts the next state
- $\mathcal{R}$ predicts the next (immediate) reward, e.g.
$$\mathcal{P}^a _{ss'}=\mathbb{P}[S'=s'|S_t=s,A_t=a]$$ 
$$\mathcal{R}^a_{s}=\mathbb{E}[R |S=s,A_t=a]$$
